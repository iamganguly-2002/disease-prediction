```python
df_main.fillna('none', inplace=True)
```

```python
df_main.head(4)
```

```python
severity_dict = dict(zip(df_severity['Symptom'].str.strip(), df_severity['weight']))
```

```python
severity_dict
```

```python
def encode_symptoms(row):
    return [severity_dict.get(sym.strip(), 0) for sym in row]
```

```python
X_encoded = df_main.iloc[:, 1:].apply(encode_symptoms, axis=1, result_type='expand')
y = df_main['Disease']
```

```python
X_encoded.head()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
y_pred_dt = dt.predict(X_test)

print("DT Report:\n", classification_report(y_test, y_pred_dt, target_names=le.classes_))
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dt,filled=True)
```

```python
import xgboost as xgb
```

```python
xgb_clf = xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
xgb_clf.fit(X_train, y_train)
```

```python
y_pred_xgb = xgb_clf.predict(X_test)

print("XGBoost Report:\n", classification_report(y_test, y_pred_xgb, target_names=le.classes_))
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential([
    Dense(64, input_dim=X_train.shape[1], activation='relu'),
    Dense(32, activation='relu'),
    Dense(len(le.classes_), activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=10, verbose=0)
```

```python
loss, acc = model.evaluate(X_test, y_test)
print("ANN Accuracy:", acc)
```

```python
def get_description(disease_name):
    row = df_description[df_description['Disease'].str.lower() == disease_name.lower()]
    if not row.empty:
        return row['Description'].values[0]
    else:
        return "No description available."
```

```python
def get_precautions(disease_name):
    row = df_precaution[df_precaution['Disease'].str.lower() == disease_name.lower()]
    if not row.empty:
        return row.iloc[0, 1:].dropna().tolist()  # All columns except Disease
    else:
        return ["No precautions available."]
```

```python
disease_name = "Typhoid"

desc = get_description(disease_name)
precautions = get_precautions(disease_name)

print("Disease:", disease_name)
print("Description:", desc)
print("Precautions:")
for i, item in enumerate(precautions, 1):
    print(f"{i}. {item}")
```

```python
predicted_index = dt.predict(X_test)[0]
predicted_disease = le.inverse_transform([predicted_index])[0]

# Fetch explanation
description = get_description(predicted_disease)
precaution_list = get_precautions(predicted_disease)

# Print result
print("Predicted Disease:", predicted_disease)
print("\nDescription:\n", description)
print("\nPrecautions:")
for p in precaution_list:
    print("", p)
```
